### Top 10 specific reasons to call 911

```python
df["reasons"] = df['title'].apply(lambda x: x.split(":")[1].strip())
df["broad_reasons"] = df['title'].apply(lambda x: x.split(":")[0].strip())
df.head()
```

```python
df["reasons"].value_counts().head(10)
```

```python
df["broad_reasons"].value_counts()
```

### Creating countplot for the Broad Reasons

```python
sns.countplot(x='broad_reasons', data=df)
```

### Convert timestamp from String to Actual timestamp

```python
# df['timeStamp'] = df['timeStamp'].apply(lambda x: pd.Timestamp(x)) 
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df.head()
```

```python
df['timeStamp'].info()
```

### Create 3 new columns in the DF for Hour, Month and Day of the Week

```python
df['hour'] = df['timeStamp'].apply(lambda x: x.hour)
```

```python
df['month'] = df['timeStamp'].apply(lambda x: x.month)
```

```python
df['weekday'] = df['timeStamp'].apply(lambda x: x.dayofweek)
df['weekday_in_str'] = df['timeStamp'].apply(lambda x: x.strftime('%A'))
```

```python
df.head()
```

```python
df['weekday'].value_counts()
```

### Plotting graph for Month

```python
sns.countplot(x='month', data=df, hue="broad_reasons")
```

### Use group by to plot the graph to identify the day wise calls from towns

```python
grouped_df = df.groupby('weekday').count()
```
