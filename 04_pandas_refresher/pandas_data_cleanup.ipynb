```python
import pandas as pd
import numpy as np
```

```python
ratings = { 'p1': [3, 5, np.NaN], 'p2':[4, np.NaN, np.NaN], 'p3':[3, 5, 4]}
ratings_df = pd.DataFrame(ratings)
ratings_df
```

### Dropping rows with NaN (null) values

```python
ratings_df.dropna() # Drops all the rows which contains NaN
```

### Filling data with some calculated values

```python
ratings_df.fillna(ratings_df['p1'].mean())
```

### Operations in Dataframe

```python
new_data = {
    "Movies": ['Star-Wars', 'Star-Wars', 'Batman', 'Batman', 'Deadpool', 'Deadpool'],
    "Viewers": ['p1', 'p2', 'p3', 'p4', 'p5', 'p6'],
    "Ratings": [10, 7, 6, 9, 8, 5]
}
new_data_df = pd.DataFrame(new_data)
new_data_df
```

```python
grouped_data_df = new_data_df.groupby('Movies')
grouped_data_df
```

```python
grouped_data_df.mean(numeric_only=True)
```

```python
grouped_data_df.describe()
```

```python
new_data_df.apply(lambda x: x + x)
```

### Creating PIVOT tables

```python
df = pd.DataFrame(
    {
        'A': ['a','b','c','d'] * 3,
        'B': ['A', 'B', 'C'] * 4,
        'C': ['P','P','P', 'Q','Q','Q'] * 2,
        'D': np.random.randint(5,20,12),
        'E': np.random.randint(2,10, 12)
    }
)
df
```

```python
pd.pivot_table(df, values='D', index=['A','B'], columns=['C'])
```

### Merging DFs similar to JOIN operation

```python
df5 = pd.DataFrame({ 'A': [1,2], 'B': [3,4]})
df6 = pd.DataFrame({ 'A': [3,1], 'D': [4,5]})
```

```python
pd.merge(df5, df6, on='A')
```
