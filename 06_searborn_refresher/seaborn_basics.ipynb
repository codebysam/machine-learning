```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_dataset = sns.load_dataset('titanic')
titanic_dataset.head()
```

### Distribution plot

```python
sns.displot(titanic_dataset['age'], kde=True)
```

### JointPlot

```python
sns.jointplot(x='age', y='fare', data=titanic_dataset)
```

### JointPlot for Regression

```python
sns.jointplot(x='age', y='fare', data=titanic_dataset, kind='reg')
```

```python
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
```

### PairPlot

```python
sns.pairplot(tips_dataset)
```

### Cat Plot

```python
exercise_dataset = sns.load_dataset('exercise')
exercise_dataset.head()
```

```python
my_plot = sns.catplot(x='time', y='pulse', hue='kind', data=exercise_dataset, col='diet')
```

### Swarmplot with Iris Dataset

```python
iris_dataset = sns.load_dataset('iris')
iris_dataset.head()
```

```python
sns.swarmplot(x='species', y='petal_length', data=iris_dataset)
```

```python
plt.plot(iris_dataset['species'], iris_dataset['petal_width'], label='Petal Width')
plt.plot(iris_dataset['species'], iris_dataset['petal_length'], label='Petal Length')
plt.plot(iris_dataset['species'], iris_dataset['sepal_width'], label='Sepal Width')
plt.plot(iris_dataset['species'], iris_dataset['sepal_length'], label='Sepal Length')
plt.legend()
```
