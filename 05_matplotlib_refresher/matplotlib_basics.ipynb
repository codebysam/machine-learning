```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(5, 10, 100)
y = x ** 4
```

### Plotting the Graph

```python
plt.plot(x, y)
plt.xlabel("x axis: x")
plt.ylabel("y axis: x^3")
```

### Plotting two graphs with legends

```python
plt.plot(x, x**2, label="square")
plt.plot(x, x**3, label="cube")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Power graph")
plt.legend()
```

### Plotting Histogram

```python
ratings = np.random.randint(1,10, 50)
rating_range = np.arange(1,11)

plt.hist(ratings, bins=rating_range, histtype='bar', rwidth=0.5)
```

### Creating StackPlot

```python
day_range = np.arange(10)

organic = np.random.randint(5,20,10)
affiliate = np.random.randint(5,20,10)
ads = np.random.randint(5,20,10)
seller = np.random.randint(5,20,10)
```

```python
plt.stackplot(day_range, organic, affiliate, ads, seller, 
              labels=["organic", "affiliate", "ads", "seller"],
              colors=["red", "blue", "green", "yellow"])
plt.legend()
```

```python
plt.plot(day_range, affiliate, label="affiliate")
plt.plot(day_range, ads, label="ads")
plt.plot(day_range, seller, label="seller")
plt.plot(day_range, organic, label="organic")
plt.xlabel("Day Range")
plt.ylabel("Sales")
plt.legend()
```

### Creating Pie Chart

```python
pie_labels = ["Day " + str(i) for i in day_range] 
pie_labels
```

```python
plt.pie(ads, labels=pie_labels, explode=(0, 0.2, 0, 0.2, 0, 0, 0, 0.2, 0, 0))
```
